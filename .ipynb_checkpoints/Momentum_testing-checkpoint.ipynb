# 澳股动能选股策略
研究显示股票过去12个月到过去1个月（t-12，t-1）的收益能预测未来一个月的收益。买过去一年收益高的股票组合能跑赢收益低的股票组合。

```python
import yfinance as yf #需要安装：‘pip install yfinance --upgrade --no-cache-dir’
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Helper Function for date alignment
def date_adjust(d):
    while d not in whole.index:
        d = d + pd.Timedelta("1 days")
    return d

def date_adjust_back(d):
    while d not in whole.index:
        d = d - pd.Timedelta("1 days")
    return d
```

```python
#asx_200为历史asx200估值成分，whole为所有股票从2006年历史数据
asx_200 = pd.read_csv("IOZ_Monthly_Holding_2010_2021.csv")
whole = pd.read_csv("2022_02_08_Updated_ASX200_Prices_2006_2021_Complete.csv")

#将日期化为index
whole['Date'] = pd.to_datetime(whole['Date'])
whole = whole.set_index('Date')
whole = whole.dropna(axis=0,thresh=21)
```

```python
# 建立回测时间线

Start = date_adjust(pd.to_datetime("2011-01-01"))
Start_list = []
for i in range((12*11+1)):
    Start_list.append(date_adjust(Start + pd.DateOffset(months=i)))
    
Backtest_start_list = []
Backtest_end_list = []

for i in range(len(Start_list)):
    Backtest_start_list.append(date_adjust(Start_list[i] - pd.DateOffset(years=1)))
    Backtest_end_list.append(date_adjust(Start_list[i] - pd.DateOffset(months=1)))

Performance_end_list = Start_list[1:]
Performance_end_list.append(pd.to_datetime('2022-02-04'))

my_dates = pd.DataFrame(zip(Backtest_start_list,Backtest_end_list,Start_list,Performance_end_list), columns=['B_start','B_end','Start','end'])
my_dates
```

```python
#计算回测每月收益
month = 12*11+1
bins = 10
size_list = []
perform_whole = np.zeros([month,bins])
for plot_num in range(month):
    Adj = whole[asx_200[str((Start_list[plot_num] + pd.offsets.MonthBegin(-1)).date())].dropna()]
    result = (Adj.loc[Backtest_end_list[plot_num],:]/ Adj.loc[Backtest_start_list[plot_num],:])-1
    performance = (Adj.loc[Performance_end_list[plot_num],:]/ Adj.loc[Start_list[plot_num],:])-1
    tickers = result.index.values

    answer = pd.DataFrame(zip(tickers,result,performance), columns=['Ticker','Returns',"Performace"])
    answer = answer.sort_values('Returns',ascending=False)

    bis = 20
    perform = np.zeros(bins)

    perform[0] = answer['Performace'][0:bis].mean()
    perform[1] = answer['Performace'][bis:bis*2].mean()
    perform[2] = answer['Performace'][bis*2:bis*3].mean()
    perform[3] = answer['Performace'][bis*3:bis*4].mean()
    perform[4] = answer['Performace'][bis*4:bis*5].mean()
    perform[5] = answer['Performace'][-bis*5:-bis*4].mean()
    perform[6] = answer['Performace'][-bis*4:-bis*3].mean()
    perform[7] = answer['Performace'][-bis*3:-bis*2].mean()
    perform[8] = answer['Performace'][-bis*2:-bis].mean()
    perform[9] = answer['Performace'][-bis:].mean()

    perform_whole[plot_num] = perform
```

```python
#计算Geomean年化收益
#高过去收益股在左边，过去收益股在右边。回测显示高过去收益能预测高未来相对收益。
result_list = [(np.prod(perform_whole[:,i]+1))**(1/11)-1 for i in range(10)]
plt.bar(np.arange(bins),result_list)
plt.title('Portfolio Annualised Return')
plt.show()
```

```python
#画图准备
start = 1.0
plot_date = np.array(my_dates['end'])
top_20 = start * (1 + perform_whole[:,0]).cumprod()
bottom_20 = start * (1 + perform_whole[:,-1]).cumprod()
```

```python
#累计10年收益
plt.plot(plot_date,top_20,label='Top_20')
plt.plot(plot_date,bottom_20,label='Bottom_20')
plt.legend()
plt.title('Cumulative Return of $1 Invested in Top/Bottom 20 Ranking Stock')
plt.show()
```
